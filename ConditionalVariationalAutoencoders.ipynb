```python
import tensorflow as tf
import numpy as np
```

```python
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets("Mnist data/", one_hot=True)
```

```python
tf.reset_default_graph()
```

```python
X = tf.placeholder(shape=[None, 28, 28,1], dtype=tf.float32, name="input")
y = tf.placeholder(shape=[None, 28,28,1], dtype=tf.float32, name="output")
z = tf.placeholder(shape=[None, 10], dtype=tf.float32, name="class")
```

```python
initializer = tf.contrib.layers.variance_scaling_initializer
```

```python
conv1 = tf.layers.conv2d(inputs=X, activation=tf.nn.relu, filters=128, kernel_initializer=initializer(), kernel_size=[5,5], padding="SAME", strides=1)
pool1 = tf.nn.max_pool(conv1,(1,2,2,1),[1,1,1,1],padding="VALID")
conv2 = tf.layers.conv2d(inputs=pool1, activation=tf.nn.relu, filters=64, kernel_initializer=initializer(), kernel_size=(5,5), padding="SAME", strides=1)
pool2 = tf.nn.max_pool(conv2, (1,2,2,1), [1,1,1,1], padding="VALID")
flatten1 = tf.layers.flatten(pool2)
dense1 = tf.layers.dense(flatten1,units=1024, activation=tf.nn.relu, kernel_initializer=initializer())
codings_mean = tf.layers.dense(inputs = dense1, units = 10, activation = None, kernel_initializer=initializer())
codings_gamma = tf.layers.dense(inputs = dense1, units = 10, activation = None, kernel_initializer=initializer())
sampling = tf.random_normal(dtype=tf.float32,shape = tf.shape(codings_gamma))
codings = codings_mean + tf.exp(0.5*codings_gamma)*sampling
```

```python
concat1 = tf.concat([codings, z], 1)
```

```python
concat1
```

```python
dense1 = tf.layers.dense(concat1, units=1024, activation=tf.nn.relu, kernel_initializer=initializer())
dense2 = tf.layers.dense(dense1, units=7*7*64, activation=tf.nn.relu, kernel_initializer=initializer())
reshape1 = tf.reshape(dense2, shape=(-1,7,7,64))
conv1T = tf.layers.conv2d_transpose(reshape1, activation=tf.nn.relu, kernel_initializer=initializer(), filters=128, kernel_size=(2,2), padding="VALID", strides=2)
conv2T = tf.layers.conv2d_transpose(conv1T, activation=None, kernel_initializer=initializer(), filters=1, kernel_size=(2,2), padding="VALID", strides=2)
generation = tf.nn.sigmoid(conv2T)
```

```python
latent_loss = 0.5 * tf.reduce_sum(tf.square(codings_mean) + tf.exp(codings_gamma)- (1 + codings_gamma))
```

```python
reconstruction_loss = tf.reduce_sum(tf.nn.sigmoid_cross_entropy_with_logits(logits = conv2T , labels=y))
```

```python
loss = latent_loss + reconstruction_loss
```

```python
optimizer = tf.train.AdamOptimizer(learning_rate=0.001)
training_op = optimizer.minimize(loss)
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
iterations = 70
batch_per_iterations = 1000
batch_size = 64
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with tf.Session() as sess:
    sess.run(init)
    for it in range(iterations):
        for _ in range(batch_per_iterations):
            true_image, im_class = mnist.train.next_batch(batch_size)
            true_image = true_image.reshape((-1,28,28,1))
            sess.run(training_op, feed_dict={X:true_image, y:true_image, z:im_class})
        if it%5 == 0:
            generated_image = sess.run(generation, feed_dict={codings:np.random.normal(size=[9, 10]), z:im_class[:9]})
            true_image, im_class = mnist.test.next_batch(batch_size)
            true_image = true_image.reshape((-1,28,28,1))
            print("Reconstruction loss:",sess.run(reconstruction_loss, feed_dict={X:true_image, y:true_image, z:im_class}))
            for i in range(9):
                plt.subplot(3,3,i+1)
                plt.imshow(generated_image[i].reshape((28,28)), cmap="gray")
                plt.axis("off")
            plt.show()
    saver.save(sess, "./model")
```
